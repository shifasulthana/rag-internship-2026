# calculate average ragas score

```python
try:

    metrics = [
        "faithfulness",
        "answer_relevancy",
        "context_precision",
        "context_recall"
    ]
    phase2_scores = {}
    phase4_scores = {}
    for metric in metrics:
        phase2_scores[
            metric
        ] = phase2_df[
            metric
        ].mean()
        phase4_scores[
            metric
        ] = phase4_df[
            metric
        ].mean()
    print(
        "Average metric scores calculated!"
    )
except Exception as e:
    print(
        f"Calculation Error: {e}"
    )
```

# Comparison Table

```python
try:

    comparison_df = pd.DataFrame({
        "Metric":[
            "Faithfulness",
            "Answer Relevancy",
            "Context Precision",
            "Context Recall"
        ],
        "Phase 2":[
            phase2_scores[
                "faithfulness"
            ],
            phase2_scores[
                "answer_relevancy"
            ],
            phase2_scores[
                "context_precision"
            ],
            phase2_scores[
                "context_recall"
            ]
        ],
        "Phase 4":[
            phase4_scores[
                "faithfulness"
            ],
            phase4_scores[
                "answer_relevancy"
            ],
            phase4_scores[
                "context_precision"
            ],
            phase4_scores[
                "context_recall"
            ]
        ]
    })
    comparison_df[
        "Improvement"
    ] = (
        comparison_df[
            "Phase 4"
        ]
        -
        comparison_df[
            "Phase 2"
        ]
    )
    print(comparison_df)
except Exception as e:
    print(
        f"Comparison Table Error: {e}"
    )
```

```python
try:

    best_metric = comparison_df.loc[
        comparison_df[
            "Improvement"
        ].idxmax()
    ]
    print("-" * 50)
    print(" MOST IMPROVED METRIC")
    print("-" * 50)
    print(
        f"Metric        : {best_metric['Metric']}"
    )
    print(
        f"Phase 2 Score : {best_metric['Phase 2']:.4f}"
    )
    print(
        f"Phase 4 Score : {best_metric['Phase 4']:.4f}"
    )
    print(
        f"Improvement   : +{best_metric['Improvement']:.4f}"
    )
    print(
    "Interpretation:"
    )
    print(
        f"{best_metric['Metric']} showed the greatest improvement "
        f"from Phase 2 to Phase 4, increasing by "
        f"{best_metric['Improvement']:.4f}."
    )
except Exception as e:
    print(
        f"Metric Analysis Error: {e}"
    )
```

```python
try:
    print(
        "\nEvaluation Summary\n"
    )
    print(
        "-" * 50
    )
    print(
        f"Most Improved Metric : {best_metric['Metric']}"
    )
    print(
        f"Improvement Value    : {best_metric['Improvement']:.4f}"
    )
    print(
        "\nHybrid Retrieval improved answer quality while maintaining strong retrieval performance."
    )
    print(
        "Context Precision decreased slightly due to the inclusion of BM25 retrieved documents."
    )
except Exception as e:
    print(
        f"Summary Error: {e}"
    )
```

# Phase 2 vs Phase 4 Comparison

## Objective

This notebook compares the baseline RAG system (Phase 2) with the Hybrid Retrieval RAG system (Phase 4) using RAGAS evaluation metrics.

## Evaluation Metrics

- Faithfulness
- Answer Relevancy
- Context Precision
- Context Recall

## Observations

- Hybrid Retrieval improved answer relevancy.
- Faithfulness remained high, indicating that answers were grounded in the retrieved context.
- Context Recall remained strong.
- Context Precision decreased slightly because BM25 retrieval introduced additional relevant context.

## Conclusion

The Phase 4 Hybrid Retrieval system produced more relevant answers while maintaining strong retrieval performance. This comparison was performed using a 2-question evaluation due to API quota limitations.
