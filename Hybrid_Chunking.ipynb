Create ChoromaDB Database

```python
try:

    import time
    import os
    import shutil

    from langchain_chroma import Chroma

    # Remove previous Phase4 DB
    if os.path.exists(CHROMA_DIR):

        shutil.rmtree(CHROMA_DIR)

        print(" Old ChromaDB removed")

    batch_size = 20

    total_chunks = len(chunks)

    print(f"Total chunks: {total_chunks}")

    db = None

    for start in range(
        0,
        total_chunks,
        batch_size
    ):

        end = min(
            start + batch_size,
            total_chunks
        )

        batch = chunks[start:end]

        print(
            f"\nProcessing batch "
            f"{start//batch_size + 1}"
            f" | chunks {start}-{end}"
        )

        try:

            if db is None:

                db = Chroma.from_documents(
                    documents=batch,
                    embedding=embeddings,
                    collection_name="wiki_rag_phase4",
                    persist_directory=CHROMA_DIR
                )

            else:

                db.add_documents(batch)

            print(
                f" Added {len(batch)} chunks"
            )

            # Wait to avoid rate limit
            time.sleep(15)

        except Exception as batch_error:

            print(
                f" Batch failed: {batch_error}"
            )

            print(
                "Waiting 65 seconds..."
            )

            time.sleep(65)

            db.add_documents(batch)

            print(
                " Retry successful"
            )

    print(
        "\n Phase 4 ChromaDB created successfully!"
    )

except Exception as e:

    print(
        f" ChromaDB Creation Error: {e}"
    )
```

BM25 Search

```python
import re

def bm25_search(
    query,
    k=5
):

    try:

        query_tokens = re.findall(
            r"\w+",
            query.lower()
        )

        scores = bm25.get_scores(
            query_tokens
        )

        ranked = sorted(
            zip(
                chunk_texts,
                scores
            ),
            key=lambda x: x[1],
            reverse=True
        )

        return ranked[:k]

    except Exception as e:

        print(
            f"BM25 Search Error: {e}"
        )

        return []
```

Hybrid Search

```python
def hybrid_search(
    query,
    vector_k=3,
    bm25_k=2
):

    try:

        vector_docs = vector_search(
            query,
            vector_k
        )

        bm25_docs = bm25_search(
            query,
            bm25_k
        )

        final_docs = []

        seen = set()

        # Prioritize vector results
        for doc in vector_docs:

            text = doc.page_content

            if text not in seen:

                final_docs.append(doc)

                seen.add(text)

        # Add BM25 results
        for text, score in bm25_docs:

            if text not in seen:

                final_docs.append(text)

                seen.add(text)

        return final_docs

    except Exception as e:

        print(
            f"Hybrid Search Error: {e}"
        )

        return []
```

```python
try:

    import os

    from groq import Groq

    from google.colab import userdata

    GROQ_API_KEY = userdata.get(
        "Groq_Key"
    )

    if not GROQ_API_KEY:

        raise ValueError(
            "Groq_Key not found in Colab Secrets"
        )

    client = Groq(
        api_key=GROQ_API_KEY
    )

    print(
        " Groq client initialized"
    )

except Exception as e:

    print(
        f" Groq Setup Error: {e}"
    )
```

```python
def ask_question(query):

    try:

        docs = hybrid_search(query)

        if len(docs) == 0:

            return (
                "No relevant documents found."
            )

        context_parts = []

        for doc in docs:

            if hasattr(
                doc,
                "page_content"
            ):

                context_parts.append(
                    doc.page_content
                )

            else:

                context_parts.append(
                    str(doc)
                )

        context = "\n\n".join(
            context_parts
        )

        prompt = f"""
You are a helpful assistant.

Answer ONLY using the provided context.

If the answer is not present
in the context, say:

'I could not find that information
in the documents.'

Context:
{context}

Question:
{query}
"""

        response = (
            client.chat.completions.create(
                model=
                "llama-3.3-70b-versatile",
                messages=[
                    {
                        "role":"user",
                        "content":prompt
                    }
                ]
            )
        )

        return (
            response
            .choices[0]
            .message
            .content
        )

    except Exception as e:

        return f"Error: {e}"
```

## Objective
The objective of Phase 4 is to improve the Retrieval-Augmented Generation (RAG) system by implementing retrieval enhancement techniques and evaluating their impact on answer quality.

---

## Implemented Improvements

### 1. Better Chunking (Technique D)
- Replaced the previous chunking strategy with an improved paragraph-aware and sentence-aware chunking approach.
- Used:
  - Chunk Size: 1800
  - Chunk Overlap: 200
- Generated 347 structured chunks from the AI knowledge base.

**Expected Benefits**
- Improved context preservation.
- Reduced information loss across chunk boundaries.
- Better retrieval quality.

### 2. Hybrid Search (Technique A)
Implemented a Hybrid Retrieval mechanism combining:

1. Vector Search (Semantic Retrieval)
   - Gemini Embeddings
   - ChromaDB Vector Store

2. BM25 Keyword Search
   - Rank-BM25 Retriever

The final retriever combines semantic similarity and keyword matching to retrieve more relevant context.

**Expected Benefits**
- Improved Context Recall.
- Improved Context Precision.
- Better handling of exact keyword queries.
