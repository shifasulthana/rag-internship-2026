```python
import os
import sys
from google.colab import drive
from google.colab import userdata

# Mount Google Drive with exception handling
try:
    drive.mount('/content/drive')
    print(" Drive mounted successfully")

    # Verify mount by checking if directory exists
    if os.path.exists('/content/drive/MyDrive'):
        print(" Drive path verified")
    else:
        print(" Warning: Drive mounted but path not found")

except Exception as e:
    print(f" Drive mount error: {e}")
    print("Please mount drive manually by clicking the folder icon in Colab sidebar")
    sys.exit(1)
```

```python
try:
    gemini_key = userdata.get("Gemini_key")
    if not gemini_key:
        raise ValueError("Gemini_key not found in Colab secrets")

    groq_key = userdata.get("Groq_Key")
    if not groq_key:
        raise ValueError("Groq_Key not found in Colab secrets")

    # Set environment variable for Gemini
    os.environ["GOOGLE_API_KEY"] = gemini_key

    print(" API Keys loaded successfully")

except ValueError as e:
    print(f" API Key error: {e}")
    sys.exit(1)

except Exception as e:
    print(f"Unexpected error loading keys: {e}")
    sys.exit(1)
```
