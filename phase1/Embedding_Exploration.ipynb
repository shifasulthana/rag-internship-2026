```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
```

```python
sentences = [
    "Artificial intelligence is transforming healthcare.",
    "Machine learning helps computers learn patterns.",
    "Deep learning uses neural networks.",
    "Hospitals use AI for disease prediction.",
    "Doctors analyze patient records.",

    "Football is a popular sport.",
    "Cricket is widely played in India.",
    "Basketball requires teamwork.",
    "Sports improve physical fitness.",
    "Athletes train every day.",

    "Python is a programming language.",
    "Java is used in enterprise applications.",
    "Software development requires problem solving.",
    "Programming involves writing code.",
    "Developers use GitHub for collaboration.",

    "Cats are popular pets.",
    "Dogs are loyal animals.",
    "Animals need food and water.",
    "Pet owners care for their animals.",
    "Veterinarians treat sick animals."
]

print(f"Total sentences: {len(sentences)}")
```

Load Embedding Model

```python
model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

print("Model loaded successfully")
```

Generate Embeddings

```python
embeddings = model.encode(sentences)

print("Embedding Shape:")
print(embeddings.shape)
```

```python
print(embeddings[0][:10])
```

```python
query = "AI helps hospitals diagnose diseases"

query_embedding = model.encode([query])

print(query_embedding.shape)
```

Similarity Search

```python
scores = cosine_similarity(
    query_embedding,
    embeddings
)[0]

print(scores)
```

Top 5 Matches

```python
results = list(zip(sentences, scores))

results = sorted(
    results,
    key=lambda x: x[1],
    reverse=True
)

for sentence, score in results[:5]:
    print(f"{score:.4f} | {sentence}")
```

next Query

```python
query2 = "pets and animal care"

query2_embedding = model.encode([query2])

scores2 = cosine_similarity(
    query2_embedding,
    embeddings
)[0]

results2 = list(zip(sentences, scores2))

results2 = sorted(
    results2,
    key=lambda x: x[1],
    reverse=True
)

for sentence, score in results2[:5]:
    print(f"{score:.4f} | {sentence}")
```
